```
import pandas as pd

df_nodes = pd.read_csv("InputFileNodes.csv")
node_list = df_nodes['id'].tolist()

print(node_list)
```

```
df_edges = pd.read_csv('InputFileEdges.csv')
edge_list = list(zip(df_edges['from'], df_edges['to']))

print("2. Edge listesi:")
print(edge_list)

```

```
edge_list = list(set(edge_list))
print("3. Tekrarı alınmış edge listesi:")
print(edge_list)
```

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(node_list)
G.add_edges_from(edge_list)

plt.figure(figsize=(7,7))
nx.draw(G, with_labels=True, node_color="lightblue")
plt.title("G – Yönsüz Ağ")
plt.show()

G_directed = nx.DiGraph()
G_directed.add_nodes_from(node_list)
G_directed.add_edges_from(edge_list)

plt.figure(figsize=(7,7))
nx.draw(G_directed, with_labels=True, node_color="lightgreen", arrows=True)
plt.title("G_directed – Yönlü Ağ")
plt.show()
```

```
degrees = dict(G.degree())
print("5. Tüm düğümlerin dereceleri:")
print(degrees)
```

```
avg_degree = sum(degrees.values()) / len(degrees)
print("6. Ortalama derece:", avg_degree)
```

```
plt.figure(figsize=(6,4))
plt.hist(list(degrees.values()))
plt.title("7. Derece Histogramı")
plt.xlabel("Derece")
plt.ylabel("Frekans")
plt.show()
```

```
deg_cent = nx.degree_centrality(G)
print("8. Degree centrality:")
print(deg_cent)
```

```
closeness = nx.closeness_centrality(G)
print("9. Closeness centrality:")
print(closeness)
```

```
betweenness = nx.betweenness_centrality(G)
print("10. Betweenness centrality:")
print(betweenness)
```

```
eigen = nx.eigenvector_centrality(G)
print("11. Eigenvector centrality:")
print(eigen)
```

```
katz = nx.katz_centrality(G, alpha=0.1, beta=1.0)
print("12. Katz centrality:")
print(katz)
```

```
pagerank = nx.pagerank(G)
print("13. PageRank:")
print(pagerank)
```

```
from networkx.algorithms.community import greedy_modularity_communities

communities = greedy_modularity_communities(G)
modularity = nx.algorithms.community.quality.modularity(G, communities)

print("14. Modularity:", modularity)
```

```
density = nx.density(G)
print("15. Density:", density)
```

```
avg_clustering = nx.average_clustering(G)
print("16. Average clustering:", avg_clustering)
```

```
if nx.is_connected(G):
    diameter = nx.diameter(G)
    print("17. Diameter:", diameter)
else:
    print("17. Ağ bağlantılı değil, diameter hesaplanamaz.")
```

```
degree_values = [d for n, d in G.degree()]
unique_degrees = sorted(set(degree_values))
degree_count = [degree_values.count(d) for d in unique_degrees]

plt.figure(figsize=(6,4))
plt.plot(unique_degrees, degree_count, marker='o')
plt.title("18. Derece Dağılımı")
plt.xlabel("Derece")
plt.ylabel("Frekans")
plt.show()
```

```
top5_bridge = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:5]
print("20. En kritik 5 KÖPRÜ düğümü (Betweenness):")
print(top5_bridge)
```

```
top5_leader = sorted(eigen.items(), key=lambda x: x[1], reverse=True)[:5]
print("21. En kritik 5 LİDER düğüm (Eigenvector):")
print(top5_leader)
```

```
in_degrees = dict(G_directed.in_degree())
out_degrees = dict(G_directed.out_degree())

plt.figure(figsize=(6,4))
plt.hist(in_degrees.values())
plt.title("22. G_directed – In-degree Dağılımı")
plt.show()

plt.figure(figsize=(6,4))
plt.hist(out_degrees.values())
plt.title("22. G_directed – Out-degree Dağılımı")
plt.show()
```

```
top3 = sorted(deg_cent.items(), key=lambda x: x[1], reverse=True)[:3]
top3_nodes = [n for n, v in top3]

G_removed = G.copy()
G_removed.remove_nodes_from(top3_nodes)

plt.figure(figsize=(7,7))
nx.draw(G_removed, with_labels=True, node_color="orange")
plt.title("23. En güçlü 3 düğüm çıkarıldıktan sonra ağ")
plt.show()

print("Ağdan çıkarılan düğümler:", top3_nodes)
```

```
top3 = sorted(deg_cent.items(), key=lambda x: x[1], reverse=True)[:3]
top3_nodes = [x[0] for x in top3]

print("24) Silinen En Güçlü 3 Düğüm:", top3_nodes)

G_removed = G.copy()
G_removed.remove_nodes_from(top3_nodes)

plt.figure(figsize=(8, 8))
nx.draw(G_removed, with_labels=True, node_size=800)
plt.title("24) En Güçlü 3 Düğüm Silindikten Sonra Yeni Ağ")
plt.show()
```
