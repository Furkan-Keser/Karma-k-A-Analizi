1.SORU

```
import pandas as pd

teams = pd.read_csv("mlb_teams.csv", encoding="utf-16")
teams = teams.loc[:, ~teams.columns.str.contains("Unnamed")]
teams.head()
```

2.SORU

```
games = pd.read_csv("mlb_games.csv", encoding="utf-16")
games = games.loc[:, ~games.columns.str.contains("Unnamed")]
games.head()

```

3.SORU

```
import networkx as nx

# Olası kolonlar
possible = ["home_team","away_team","home","away","team1","team2","teamA","teamB"]

cols = [c for c in possible if c in games.columns]
team1, team2 = cols[0], cols[1]

edges = list(zip(games[team1], games[team2]))

G = nx.Graph()
G.add_edges_from(edges)   # tekrar eden kenarlar Graph tarafından otomatik silinir

print("Toplam düğüm:", G.number_of_nodes())
print("Toplam kenar:", G.number_of_edges())

```

4.SORU

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True, node_size=400, font_size=8)
plt.show()
```

5.SORU

```
G_directed = nx.DiGraph()
G_directed.add_edges_from(edges)

plt.figure(figsize=(10, 10))
nx.draw(G_directed, with_labels=True, node_size=400, font_size=8)
plt.show()
```

6.SORU

```
degrees = dict(G.degree())
degrees
```

7.SORU

```
avg_degree = sum(degrees.values()) / G.number_of_nodes()
avg_degree
```

8.SORU

```
plt.hist(list(degrees.values()))
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.title("Degree Histogram")
plt.show()
```

9.SORU

```
degree_centrality = nx.degree_centrality(G)
degree_centrality
```

10.SORU

```
closeness_centrality = nx.closeness_centrality(G)
closeness_centrality
```

11.SORU

```
betweenness_centrality = nx.betweenness_centrality(G)
betweenness_centrality
```

12.SORU

```
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)
eigenvector_centrality
```

13.SORU

```
katz_centrality = nx.katz_centrality(G, alpha=0.005, beta=1.0, max_iter=2000)
katz_centrality
```

14.SORU

```
pagerank = nx.pagerank(G)
pagerank
```

15.SORU

```
import networkx.algorithms.community as nx_comm

communities = nx_comm.greedy_modularity_communities(G)
modularity = nx_comm.modularity(G, communities)
modularity
```

16.SORU

```
density = nx.density(G)
density
```

17.SORU

```
average_clustering = nx.average_clustering(G)
average_clustering
```

18.SORU

```
largest_cc = max(nx.connected_components(G), key=len)
G_lcc = G.subgraph(largest_cc)

diameter = nx.diameter(G_lcc)
diameter
```
